```python
from lights import *
import structuredlight as sl
import imageio,os,glob
import matplotlib.pyplot as plt
import cv2

uncalibrate_dir = os.path.join(os.getcwd(),"calibrate","gamma","uncorrection_captures")
calibrate_dir = os.path.join(os.getcwd(),"calibrate","gamma","correction_captures")
uncalibrate_img_dir = glob.glob(os.path.join(uncalibrate_dir,"*.png"))
calibrate_img_dir = glob.glob(os.path.join(calibrate_dir,"*.png"))
```

```python
uncalibrate_img_list = []
calibrate_img_list = []
for uncalibrate_img,calibrate_img in zip(uncalibrate_img_dir,calibrate_img_dir):
    uncalibrate_img_list.append(cv2.imread(uncalibrate_img,cv2.IMREAD_GRAYSCALE))
    calibrate_img_list.append(cv2.imread(calibrate_img,cv2.IMREAD_GRAYSCALE))
uncalibrate_img_list = np.stack(uncalibrate_img_list,axis=0)
calibrate_img_list = np.stack(calibrate_img_list,axis=0)
uncalibrate_std_list = np.std(uncalibrate_img_list,axis=(1,2))
calibrate_std_list = np.std(calibrate_img_list,axis=(1,2))
```
